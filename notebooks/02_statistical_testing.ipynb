# 🧪 02 — Statistical Testing
## Bank Marketing Campaign

**Objective:** Test whether conversion rate differences across segments are statistically significant.

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')
from src.utils import load_data

sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = load_data('../data/bank-additional-full.csv')
df['subscribed'] = (df['y'] == 'yes').astype(int)
print(f'Overall conversion rate: {df["subscribed"].mean():.4f}')
```

## 1. Chi-Square Test — Subscription vs. Categorical Variables

H0: The variable is independent of subscription.
H1: There is a relationship between the variable and subscription.

```python
categorical_vars = ['job', 'marital', 'education', 'contact', 'month', 'poutcome', 'housing', 'loan']

results = []
for var in categorical_vars:
    contingency = pd.crosstab(df[var].dropna(), df['subscribed'])
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency)
    results.append({'Variable': var, 'Chi2': chi2, 'p-value': p_value, 'dof': dof, 'Significant (5%)': p_value < 0.05})

chi_results = pd.DataFrame(results).sort_values('p-value')
chi_results
```

## 2. T-Test — Numerical Variables by Subscription

```python
numerical_vars = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']

t_results = []
for var in numerical_vars:
    group_yes = df[df['subscribed'] == 1][var].dropna()
    group_no = df[df['subscribed'] == 0][var].dropna()
    t_stat, p_value = stats.ttest_ind(group_yes, group_no)
    t_results.append({
        'Variable': var,
        'Mean (Yes)': group_yes.mean(),
        'Mean (No)': group_no.mean(),
        'Difference': group_yes.mean() - group_no.mean(),
        't-statistic': t_stat,
        'p-value': p_value,
        'Significant (5%)': p_value < 0.05
    })

t_df = pd.DataFrame(t_results).sort_values('p-value')
t_df
```

## 3. Conversion Rate Comparison by Contact Type

Is cellular contact significantly better than telephone?

```python
cellular = df[df['contact'] == 'cellular']['subscribed']
telephone = df[df['contact'] == 'telephone']['subscribed']

print(f'Cellular conversion: {cellular.mean():.4f} (n={len(cellular)})')
print(f'Telephone conversion: {telephone.mean():.4f} (n={len(telephone)})')

t_stat, p_value = stats.ttest_ind(cellular, telephone)
print(f'\nt-statistic: {t_stat:.4f}')
print(f'p-value: {p_value:.6f}')
print(f'Significant at 5%: {p_value < 0.05}')
```

## 4. Effect of Previous Campaign Outcome

```python
# Conversion by previous outcome
for outcome in df['poutcome'].dropna().unique():
    group = df[df['poutcome'] == outcome]['subscribed']
    print(f'{outcome}: conversion = {group.mean():.4f} (n={len(group)})')
```
