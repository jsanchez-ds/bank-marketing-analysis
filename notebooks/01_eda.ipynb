# 📊 01 — Exploratory Data Analysis
## Bank Marketing Campaign

**Objective:** Understand the dataset structure, distributions, and relationships to inform modeling decisions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')
from src.utils import load_data, plot_target_distribution, plot_conversion_by_feature

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
%matplotlib inline
```

## 1. Data Loading & First Look

```python
df = load_data('../data/bank-additional-full.csv')
print(f'Dataset shape: {df.shape}')
print(f'\nColumn types:\n{df.dtypes}')
df.head()
```

```python
# Missing values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
# Basic statistics
df.describe()
```

## 2. Target Distribution

```python
plot_target_distribution(df, 'y')
```

## 3. Numerical Features

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
for i, col in enumerate(numerical_cols[:9]):
    ax = axes[i // 3, i % 3]
    df[col].hist(bins=30, ax=ax, color='#3498db', edgecolor='white')
    ax.set_title(col)
plt.tight_layout()
plt.savefig('../figures/numerical_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Categorical Features

```python
categorical_cols = df.select_dtypes(include=['object']).columns.drop('y').tolist()

fig, axes = plt.subplots(3, 3, figsize=(16, 12))
for i, col in enumerate(categorical_cols[:9]):
    ax = axes[i // 3, i % 3]
    df[col].value_counts().plot(kind='bar', ax=ax, color='#2ecc71')
    ax.set_title(col)
    ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.savefig('../figures/categorical_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Bivariate Analysis — Conversion Rate by Feature

```python
for feat in ['job', 'education', 'marital', 'contact', 'month', 'poutcome']:
    plot_conversion_by_feature(df, feat)
```

## 6. Correlation Matrix

```python
df_encoded = df.copy()
df_encoded['y_num'] = (df_encoded['y'] == 'yes').astype(int)
corr = df_encoded[numerical_cols + ['y_num']].corr()

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', center=0, ax=ax)
ax.set_title('Correlation Matrix')
plt.tight_layout()
plt.savefig('../figures/correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Feature Engineering

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 30, 40, 50, 60, 100], labels=['<30', '30-40', '40-50', '50-60', '60+'])
df['has_housing_loan'] = (df['housing'] == 'yes').astype(int)
df['has_personal_loan'] = (df['loan'] == 'yes').astype(int)
df['contact_cellular'] = (df['contact'] == 'cellular').astype(int)

plot_conversion_by_feature(df, 'age_group')
```
